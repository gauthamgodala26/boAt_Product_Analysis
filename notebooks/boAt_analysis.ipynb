```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/Airdopes_reviews.csv')

# Basic cleanup
df['review_text'] = df['review_text'].astype(str).str.lower()

# Simple sentiment tagging
positive_kw = ['good', 'great', 'love', 'excellent', 'amazing']
negative_kw = ['bad', 'poor', 'disappointed', 'worst', 'terrible']

def tag_sentiment(text):
    if any(k in text for k in positive_kw):
        return 'Positive'
    if any(k in text for k in negative_kw):
        return 'Negative'
    return 'Neutral'

df['sentiment'] = df['review_text'].apply(tag_sentiment)

# Feature tagging
def extract_feature(text):
    if 'battery' in text:
        return 'Battery'
    if 'sound' in text:
        return 'Sound Quality'
    if 'comfort' in text:
        return 'Comfort'
    if 'design' in text:
        return 'Design'
    return 'Other'

df['feature_tag'] = df['review_text'].apply(extract_feature)

# Results
sentiment_summary = df['sentiment'].value_counts()
feature_summary = df['feature_tag'].value_counts()

sentiment_summary, feature_summary
```
